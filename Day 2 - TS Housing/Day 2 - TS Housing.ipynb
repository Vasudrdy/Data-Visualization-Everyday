```python
plt.figure(figsize=(20, 15))
sns.barplot(data=df, x='Districts', y='Houses Allotted')

plt.title("District vs Houses Allotted", fontsize=20)
plt.xlabel("Districts", fontsize=16)
plt.ylabel("Houses Allotted", fontsize=16)
plt.xticks(rotation=45, ha='right')      
plt.show()
```

```python
plt.figure(figsize=(20, 15))
sns.barplot(data=df, x='Districts', y='Houses Allotted Rural')

plt.title("District vs Houses Allotted Rural", fontsize=20)
plt.xlabel("Districts", fontsize=16)
plt.ylabel("Houses Allotted", fontsize=16)
plt.xticks(rotation=45, ha='right')      
plt.show()
```

```python
plt.figure(figsize=(20, 15))
sns.barplot(data=df, x='Districts', y='Houses Allotted Urban')

plt.title("District vs Houses Allotted Urban", fontsize=20)
plt.xlabel("Districts", fontsize=16)
plt.ylabel("Houses Allotted", fontsize=16)
plt.xticks(rotation=45, ha='right')      
plt.show()
```

```python
plt.figure(figsize=(20, 15))
sns.barplot(data=df, x='Districts', y='Houses Sanctioned')

plt.title("District vs Houses Sanctioned", fontsize=20)
plt.xlabel("Districts", fontsize=16)
plt.ylabel("Houses Allotted", fontsize=16)
plt.xticks(rotation=45, ha='right')      
plt.show()
```
