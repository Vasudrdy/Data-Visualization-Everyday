```python
plt.figure(figsize=(20, 15))
sns.scatterplot(data=df, x='Commodity', y='No of Farmers', hue='District')

plt.title("Commodity vs No of Farmers by Location", fontsize=20)
plt.xlabel("Commodity", fontsize=16)
plt.ylabel("No of Farmers", fontsize=16)
plt.xticks(rotation=45, ha='right') 
plt.legend(title='Location')         
plt.show()
```
