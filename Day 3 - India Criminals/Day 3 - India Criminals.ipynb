```python
df = pd.read_csv('NCRB_CII-2019_Table_19B.2 - Actual total.csv')
#swaping file with cleaned data file with rows of 0nly actual total count
```

```python
plt.figure(figsize=(20, 15))
df_melted = df.melt(id_vars='City',value_vars=['Total   - Male', 'Total   - Female'],var_name='Gender',value_name='Total')
sns.barplot(data=df_melted, x='City', y='Total', hue='Gender')
plt.title("City vs Total criminals on gender ", fontsize=20)
plt.xlabel("City", fontsize=16)
plt.ylabel("Total criminals on gender", fontsize=16)
plt.xticks(rotation=45, ha='right')      
plt.show()
```

```python
df = pd.read_csv('NCRB_CII-2019_Table_19B.2 - Total.csv')
#swaping file with cleaned data file with rows of Total criminals different ages
```

```python
plt.figure(figsize=(20, 15))
df_melted = df.melt(id_vars='City',value_vars=[' Juveniles Apprehended - Total','18 and above and below 30 years - Total','30 and above and below 45 years - Total','45 and above and below 60 years - Total','60 years and above - Total'],var_name='Age', value_name='Total')
sns.barplot(data=df_melted, x='City', y='Total', hue='Age')
plt.title("City vs Age group of female criminals", fontsize=20)
plt.xlabel("City", fontsize=16)
plt.ylabel("Age group of female criminals", fontsize=16)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
